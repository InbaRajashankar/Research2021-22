```python
def deleteNull(df):
    for ind in df.index:
        if df.isnull()['FOI_TEXT'][ind] == True:
            df.drop(labels=[ind],axis=0,inplace=True)
```

```python
def containsWord(s, w):
    return (' ' + str(w) + ' ') in (' ' + str(s) + ' ')
```

```python
def selectCohortOne(df):
    dfLWS = df[df['DEVICE_REPORT_PRODUCT_CODE']=='LWS']
    dfMRM = df[df['DEVICE_REPORT_PRODUCT_CODE']=='MRM']
    dfNIK = df[df['DEVICE_REPORT_PRODUCT_CODE']=='NIK']
    dfNVY = df[df['DEVICE_REPORT_PRODUCT_CODE']=='NVY']
    
    df = pd.concat([dfLWS, dfMRM, dfNIK, dfNVY])
```
